The Problem Dampener is a reactor-mounted module that lets the reactor safety systems tolerate a single bad level in what would otherwise be a safe report. It's like the bad level never happened!

````python
#day2 => initializer-cell

def read_input(input: str) -> list:
    return [list(map(int, line.split())) for line in input.splitlines() if line]

def test_report(report: list) -> bool:
    diffs = [report[i + 1] - report[i] for i in range(len(report) - 1)]
    return all(abs(diff) <= 3 for diff in diffs) and (all(diff > 0 for diff in diffs) or all(diff < 0 for diff in diffs))

def puzzle1(input: str) -> int:
    reports = read_input(input)
    safe = 0
    for report in reports:
        if test_report(report):
            safe += 1
    return safe

def puzzle2(input: str) -> int:
    reports = read_input(input)
    safe = 0
    for report in reports:
        delete = -1
        while delete < len(report):
            copy = report.copy()
            if delete >= 0:
                copy.pop(delete)
            if test_report(copy):
                safe += 1
                break
            delete += 1
    return safe

input_demo_1 = '''\
7 6 4 2 1
1 2 7 8 9
9 7 6 2 1
1 3 2 4 5
8 6 4 4 1
1 3 6 7 9
'''
assert puzzle1(input_demo_1) == 2
assert puzzle2(input_demo_1) == 4

input = readInput(2024, 2)
result = puzzle1(input)
# Day 2, puzzle 1: 564
print(f"Puzzle 1, safe reports {result}")
result = puzzle2(input)
# Day 2, puzzle 2: 604
print(f"Puzzle 2, safe reports with a deletion {result}")
````

# Part Two

As you scan through the corrupted memory, you notice that some of the conditional statements are also still intact. If you handle some of the uncorrupted conditional statements in the program, you might be able to get an even more accurate result.

There are two new instructions you'll need to handle:

The `do()` instruction enables future mul instructions.
The `don't()` instruction disables future mul instructions.
Only the most recent `do()` or `don't()` instruction applies. At the beginning of the program, mul instructions are enabled.

For example:
```
xmul(2,4)&mul[3,7]!^don't()_mul(5,5)+mul(32,64](mul(11,8)undo()?mul(8,5))
```

This corrupted memory is similar to the example from before, but this time the `mul(5,5)` and `mul(11,8)` instructions are disabled because there is a `don't()` instruction before them. The other mul instructions function normally, including the one at the end that gets re-enabled by a `do()` instruction.

Handle the new instructions; what do you get if you add up all of the results of just the enabled multiplications?

````python
#day3 => initializer-cell
import re

RE_MUL = re.compile(r'mul\((\d{1,3}),(\d{1,3})\)')
RE_MUL_OPS = re.compile(r'''(mul\((\d{1,3}),(\d{1,3})\))|(do(?:n't)?\(\))''')

def puzzle1(input: str) -> int:
    muls = [list(map(int, x.groups())) for x in RE_MUL.finditer(input)]
    result = 0
    for x, y in muls:
        result += x * y
    return result

def puzzle2(input: str) -> int:
    muls = [x.group() for x in RE_MUL_OPS.finditer(input)]
    result = 0
    active = True
    for mul in muls:
        if mul == 'do()':
            active = True
        elif mul == 'don\'t()':
            active = False
        elif active:
            x, y = map(int, RE_MUL.match(mul).groups())
            result += x * y
    return result

input_demo_1 = '''\
xmul(2,4)%&mul[3,7]!@^do_not_mul(5,5)+mul(32,64]then(mul(11,8)mul(8,5))
'''
assert puzzle1(input_demo_1) == 161
input_demo_2 = '''\
xmul(2,4)&mul[3,7]!^don't()_mul(5,5)+mul(32,64](mul(11,8)undo()?mul(8,5))
'''
assert puzzle2(input_demo_2) == 48

input = readInput(2024, 3)
result = puzzle1(input)
# Day 2, puzzle 1: 184576302
print(f"Puzzle 1, muls {result}")
result = puzzle2(input)
# Day 2, puzzle 2: 118173507
print(f"Puzzle 2, muls {result}")
````

````python
#day4 => initializer-cell
import re

def read_input(input: str) -> list:
    return [line for line in input.splitlines() if line]

def findXmas(line: str) -> int:
    return len(list(re.finditer('XMAS', line))) + len(list(re.finditer('SAMX', line)))

def puzzle1(input: str) -> int:
    grid = read_input(input)
    result = 0
    width = len(grid[0])
    height = len(grid)
    maxdiag = max(width, height)
    # lines
    for line in grid:
        result+= findXmas(line)
    # columns
    for i in range(width):
        col = ''.join([line[i] for line in grid])
        result+= findXmas(col)
    # diagonal \
    for i in range(width):
        line = ''.join([grid[j][i + j] for j in range(0, maxdiag) if j < height and i + j < width])
        result+= findXmas(line)
    for i in range(1, height):
        line = ''.join([grid[i + j][j] for j in range(0, maxdiag) if j < width and i + j < height])
        result+= findXmas(line)
    # diagonal /
    for i in range(width):
        line = ''.join([grid[j][i - j] for j in range(0, maxdiag) if j < height and i - j >= 0])
        result+= findXmas(line)
    for i in range(1, height):
        line = ''.join([grid[i + j][width - 1 - j] for j in range(0, maxdiag) if i + j < height and width - 1 - j >= 0])
        result+= findXmas(line)
    return result

def puzzle2(input: str) -> int:
    grid = read_input(input)
    result = 0
    width = len(grid[0])
    height = len(grid)
    maxdiag = max(width, height)
    # lines
    for i in range(1, height - 1):
        for j in range(1, width - 1):
            if grid[i][j] == 'A':
                setBackSlash = {grid[i - 1][j - 1], grid[i + 1][j + 1]}
                setSlash = {grid[i - 1][j + 1], grid[i + 1][j - 1]}
                if 'M' in setBackSlash and 'S' in setBackSlash and 'M' in setSlash and 'S' in setSlash:
                    result += 1
    return result

input_demo_1 = '''\
MMMSXXMASM
MSAMXMSMSA
AMXSXMAAMM
MSAMASMSMX
XMASAMXAMM
XXAMMXXAMA
SMSMSASXSS
SAXAMASAAA
MAMMMXMMMM
MXMXAXMASX
'''
assert puzzle1(input_demo_1) == 18
assert puzzle2(input_demo_1) == 9

input = readInput(2024, 4)
result = puzzle1(input)
# Day 4, puzzle 1: 2547
print(f"Puzzle 1, xmas {result}")
result = puzzle2(input)
# Day 4, puzzle 2: 1939
print(f"Puzzle 2, x-mas {result}")
````

````python
#day7 => initializer-cell
import re
import itertools

def read_input(input: str) -> list:
    lines = [line for line in input.splitlines() if line]
    inputs = []
    for line in lines:
        match = re.match(r'(\d+): (.*)', line)
        if not match:
            continue
        test_value = int(match.group(1))
        equations = match.group(2).split()
        inputs.append((test_value, equations))
    return inputs

def generate_expressions(digits: list, operators: list) -> list:
  """Generates all possible expressions with + and * between digits.

  Args:
    digits: A list of digits.

  Returns:
    A list of strings representing all possible expressions.
  """

  if len(digits) < 2:
    return [str(digits[0])] if digits else []

  expressions = []
  for ops in itertools.product(operators, repeat=len(digits) - 1):
    expression = [int(digits[0])]
    for i, op in enumerate(ops):
      expression.append(op)
      expression.append(int(digits[i + 1]))
    expressions.append(expression)
  return expressions

def calculate_expression(expression: list) -> int:
    result = expression[0]
    for i in range(1, len(expression), 2):
        operator = expression[i]
        operand = int(expression[i + 1])
        if operator == '+':
            result += operand
        elif operator == '*':
            result *= operand
        elif operator == '||':
            result = int(str(result) + str(operand))
    return result

def puzzle1(input: str) -> int:
    inputs = read_input(input)
    total = 0
    for test_value, equations in inputs:
        expressions = generate_expressions(equations, ['+', '*'])
        for expression in expressions:
            if calculate_expression(expression) == test_value:
                total += test_value
                break
    print('total:', total)
    return total

def puzzle2(input: str) -> int:
    inputs = read_input(input)
    total = 0
    for test_value, equations in inputs:
        expressions = generate_expressions(equations, ['+', '*', '||'])
        for expression in expressions:
            if calculate_expression(expression) == test_value:
                total += test_value
                break
    print('total:', total)
    return total


input_demo_1 = '''\
190: 10 19
3267: 81 40 27
83: 17 5
156: 15 6
7290: 6 8 6 15
161011: 16 10 13
192: 17 8 14
21037: 9 7 18 13
292: 11 6 16 20
'''
assert puzzle1(input_demo_1) == 3749
assert puzzle2(input_demo_1) == 11387

input = readInput(2024, 7)
result = puzzle1(input)
# Day 7, puzzle 1: 12940396350192
print(f"Puzzle 1: {result}")
result = puzzle2(input)
# Day 7, puzzle 2: 106016735664498
print(f"Puzzle 2: {result}")
````

````python
def array_to_graph(array):
    """Transforms a 2D array into a graph representation.

    Args:
        array: The 2D array to convert.

    Returns:
        A dictionary representing the graph where keys are nodes (cell coordinates)
        and values are lists of neighboring nodes.
    """

    graph = {}
    rows, cols = len(array), len(array[0])

    for i in range(rows):
        for j in range(cols):
            node = (i, j)  # Cell coordinates as the node
            neighbors = []

            # Check for neighbors (up, down, left, right)
            for dx, dy in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                ni, nj = i + dx, j + dy
                if 0 <= ni < rows and 0 <= nj < cols and array[ni][nj] == array[i][j] + 1:
                    neighbors.append((ni, nj))

            graph[node] = neighbors

    return graph

def find_all_paths(graph, start, end, path=[]):
    """Finds all paths from start to end in a graph using depth-first search."""

    path = path + [start]  # Add current node to path
    if start == end:
        return [path]  # Found the end node, return path

    if start not in graph:
        return []  # Node not in graph, no paths

    paths = []
    for node in graph[start]:
        if node not in path:  # Avoid cycles
            newpaths = find_all_paths(graph, node, end, path)
            for newpath in newpaths:
                paths.append(newpath)  # Add found paths

    return paths

# Example Usage
array = [[0, 1, 2, 3], [1, 2, 3, 4], [8, 7, 6, 5], [9, 8, 7, 6]]
graph = array_to_graph(array)
print(graph)

# Convert node values to tuples for compatibility with graph keys
start_node = (0, 0)  # Node with value 1
end_node = (2, 2)  # Node with value 9

all_paths = find_all_paths(graph, start_node, end_node)
print(all_paths)
````

````python
#day16 => initializer-cell

X = 0
Y = 1
DIRECTIONS = [(-1, 0), (0, -1), (0, 1), (1, 0)]

def is_valid(x: int, y: int, width: int, height: int) -> bool:
    return (x >= 0) and (x < width) and (y >= 0) and (y < height)

def bfs(ram, src: tuple, dst: tuple):
    # check source and destination cell of the matrix have value 1
    width = len(ram[0])
    height = len(ram)
    if ram[src[Y]][src[X]] == '#' or ram[dst[Y]][dst[X]] == '#':
        return -1
    visited = [[False for _ in range(width)] for _ in range(height)]
    # Mark the source cell as visited
    visited[src[Y]][src[X]] = True
    # Create a queue for BFS
    q = []
    # Distance of source cell is 0
    q.append((src, 0)) # Enqueue source cell
    # Do a BFS starting from source cell
    while q:
        pt, dist = q.pop(0) # Dequeue the front cell
        # If we have reached the destination cell, we are done
        if pt[:2] == dst:
            return dist
        # Otherwise enqueue its adjacent cells
        for dir_y, dir_x in DIRECTIONS:
            row = pt[Y] + dir_y
            col = pt[X] + dir_x
            # if adjacent cell is valid, has path and not visited yet, enqueue it.
            if (is_valid(row, col, width, height) and ram[row][col] != '#' and not visited[row][col]):
                visited[row][col] = True
                q.append(((col, row, pt[2], pt[3]), dist + 1))
    # Return -1 if destination cannot be reached
    return -1

def read_input(input: str) -> tuple:
    return [list(line) for line in input.split('\n') if line]

def print_maze(maze: list):
    for line in maze:
        print(''.join(line))

def find(maze: list, target: str) -> tuple:
    for y, line in enumerate(maze):
        for x, char in enumerate(line):
            if char == target:
                return (x, y)
    return None

def puzzle1(input: str) -> int:
    maze = read_input(input)
    print_maze(maze)
    s_x, s_y = find(maze, 'S')
    e_x, e_y = find(maze, 'E')
    print(f'reindeer start at ({s_x}, {s_y}) and end at ({e_x}, {e_y})')
    x, y, x_dir, y_dir, score = s_x, s_y, 1, 0, 0
    score = bfs(maze, (x, y, x_dir, y_dir), (e_x, e_y))
    print('score:', score)
    return score

def puzzle2(input: str) -> int:
    maze = read_input(input)
    print_maze(maze)
    s_x, s_y = find(maze, 'S')
    e_x, e_y = find(maze, 'E')
    print(f'reindeer start at ({s_x}, {s_y}) and end at ({e_x}, {e_y})')
    x, y, x_dir, y_dir, score = s_x, s_y, 1, 0, 0
    print('score:', score)
    return score

input_demo_1 = '''\
###############
#.......#....E#
#.#.###.#.###.#
#.....#.#...#.#
#.###.#####.#.#
#.#.#.......#.#
#.#.#####.###.#
#...........#.#
###.#.#####.#.#
#...#.....#.#.#
#.#.#.###.#.#.#
#.....#...#.#.#
#.###.#.#.#.#.#
#S..#.....#...#
###############
'''

input_demo_2 = '''\
#################
#...#...#...#..E#
#.#.#.#.#.#.#.#.#
#.#.#.#...#...#.#
#.#.#.#.###.#.#.#
#...#.#.#.....#.#
#.#.#.#.#.#####.#
#.#...#.#.#.....#
#.#.#####.#.###.#
#.#.#.......#...#
#.#.###.#####.###
#.#.#...#.....#.#
#.#.#.#####.###.#
#.#.#.........#.#
#.#.#.#########.#
#S#.............#
#################
'''

assert puzzle1(input_demo_1) == 7036
assert puzzle1(input_demo_2) == 11048
assert puzzle2(input_demo_1) == 0

input = readInput(2024, 16)
result = puzzle1(input)
# Day 15, puzzle 1: ???
print(f"Puzzle 1: {result}")
result = puzzle2(input)
# Day 15, puzzle 2: ???
print(f"Puzzle 2: {result}")
````

The computer knows eight instructions, each identified by a 3-bit number (called the instruction's opcode). Each instruction also reads the 3-bit number after it as an input; this is called its operand.

A number called the instruction pointer identifies the position in the program from which the next opcode will be read; it starts at `0`, pointing at the first 3-bit number in the program. Except for jump instructions, the instruction pointer increases by `2` after each instruction is processed (to move past the instruction's opcode and its operand). If the computer tries to read an opcode past the end of the program, it instead halts.

So, the program `0,1,2,3` would run the instruction whose opcode is `0` and pass it the operand `1`, then run the instruction having opcode `2` and pass it the operand `3`, then halt.

There are two types of operands; each instruction specifies the type of its operand. The value of a literal operand is the operand itself. For example, the value of the literal operand `7` is the number `7`. The value of a combo operand can be found as follows:

* Combo operands `0` through `3` represent literal values `0` through `3`.
* Combo operand `4` represents the value of register `A`.
* Combo operand `5` represents the value of register `B`.
* Combo operand `6` represents the value of register `C`.
* Combo operand `7` is reserved and will not appear in valid programs.

The eight instructions are as follows:

The `adv` instruction (opcode `0`) performs division. The numerator is the value in the A register. The denominator is found by raising 2 to the power of the instruction's combo operand. (So, an operand of 2 would divide A by 4 (2^2); an operand of 5 would divide A by 2^B.) The result of the division operation is truncated to an integer and then written to the A register.

The `bxl` instruction (opcode `1`) calculates the bitwise XOR of register B and the instruction's literal operand, then stores the result in register B.

The `bst` instruction (opcode `2`) calculates the value of its combo operand modulo 8 (thereby keeping only its lowest 3 bits), then writes that value to the B register.

The `jnz` instruction (opcode `3`) does nothing if the A register is 0. However, if the A register is not zero, it jumps by setting the instruction pointer to the value of its literal operand; if this instruction jumps, the instruction pointer is not increased by 2 after this instruction.

The `bxc` instruction (opcode `4`) calculates the bitwise XOR of register `B` and register `C`, then stores the result in register `B`. (For legacy reasons, this instruction reads an operand but ignores it.)

The `out` instruction (opcode `5`) calculates the value of its combo operand modulo 8, then outputs that value. (If a program outputs multiple values, they are separated by commas.)

The `bdv` instruction (opcode `6`) works exactly like the adv instruction except that the result is stored in the `B` register. (The numerator is still read from the `A` register.)

The `cdv` instruction (opcode `7`) works exactly like the adv instruction except that the result is stored in the `C` register. (The numerator is still read from the `A` register.)

Here are some examples of instruction operation:

* If register `C` contains `9`, the program `2,6` would set register `B` to `1`.
* If register `A` contains `10`, the program `5,0,5,1,5,4` would output `0,1,2`.
* If register `A` contains `2024`, the program `0,1,5,4,3,0` would output `4,2,5,6,7,7,7,7,3,1,0` and leave `0` in register `A`.
* If register `B` contains `29`, the program `1,7` would set register `B` to `26`.
* If register `B` contains `2024` and register `C` contains `43690`, the program `4,0` would set register `B` to `44354`.

````python
#day18 => initializer-cell
# bsf algorithm from https://www.geeksforgeeks.org/shortest-path-in-a-binary-maze/

X = 0
Y = 1
DIRECTIONS = [(-1, 0), (0, -1), (0, 1), (1, 0)]

def is_valid(x: int, y: int, width: int, height: int) -> bool:
    return (x >= 0) and (x < width) and (y >= 0) and (y < height)

def bfs(ram, src: tuple, dst: tuple):
    # check source and destination cell of the matrix have value 1
    width = len(ram[0])
    height = len(ram)
    if ram[src[Y]][src[X]] == '#' or ram[dst[Y]][dst[X]] == '#':
        return -1
    visited = [[False for _ in range(width)] for _ in range(height)]
    # Mark the source cell as visited
    visited[src[Y]][src[X]] = True
    # Create a queue for BFS
    q = []
    # Distance of source cell is 0
    q.append((src, 0)) # Enqueue source cell
    # Do a BFS starting from source cell
    while q:
        pt, dist = q.pop(0) # Dequeue the front cell
        # If we have reached the destination cell, we are done
        if pt == dst:
            return dist
        # Otherwise enqueue its adjacent cells
        for dir_y, dir_x in DIRECTIONS:
            row = pt[Y] + dir_y
            col = pt[X] + dir_x
            # if adjacent cell is valid, has path and not visited yet, enqueue it.
            if (is_valid(row, col, width, height) and ram[row][col] == '.' and not visited[row][col]):
                visited[row][col] = True
                q.append(((col, row), dist + 1))
    # Return -1 if destination cannot be reached
    return -1

def read_input(input: str) -> list:
    return [tuple(map(int, line.split(','))) for line in input.split('\n') if line]

def create_ram(width: int, height: int) -> list:
    return [['.' for _ in range(width)] for _ in range(height)]

def init_ram(ram: list, positions: list, limit: int) -> None:
    for x, y in positions[:limit]:
        ram[y][x] = '#'

def print_ram(ram: list):
    for line in ram:
        print(''.join(line))

def puzzle1(input: str, ram_size: int, limit: int) -> int:
    positions = read_input(input)
    ram = create_ram(ram_size, ram_size)
    init_ram(ram, positions, limit)
    print_ram(ram)
    src = (0, 0)
    dst = (ram_size - 1, ram_size - 1)
    distance = bfs(ram, src, dst)
    print(f'distance: {distance}')
    return distance

def puzzle2(input: str, ram_size: int, limit: int) -> str:
    positions = read_input(input)
    ram = create_ram(ram_size, ram_size)
    init_ram(ram, positions, limit)
    print_ram(ram)
    src = (0, 0)
    dst = (ram_size - 1, ram_size - 1)
    for x, y in positions[limit:]:
        ram[y][x] = '#'
        if bfs(ram, src, dst) == -1:
            print(f'first blocking byte: {x},{y}')
            return f'{x},{y}'
    return ''

input_demo_1 = '''\
5,4
4,2
4,5
3,0
2,1
6,3
2,4
1,5
0,6
3,3
2,6
5,1
1,2
5,5
2,5
6,5
1,4
0,4
6,4
1,1
6,1
1,0
0,5
1,6
2,0
'''

assert puzzle1(input_demo_1, 7, 12) == 22
assert puzzle2(input_demo_1, 7, 12) == '6,1'

input = readInput(2024, 18)
result = puzzle1(input, 71, 1024)
# Day 18, puzzle 1: 232
print(f"Puzzle 1: {result}")
result = puzzle2(input, 71, 1024)
# Day 18, puzzle 2: ???
print(f"Puzzle 2: {result}")
````

````python
#day19 => initializer-cell
from collections import defaultdict, deque

class Node(object):
    def __init__(self, value):
        self.value = value
        self.next_states = []
        self.fail_state = 0
        self.output = []

class AhoCorasick(object):

    def __init__(self, keywords):
        """ creates a trie of keywords, then sets fail transitions """
        self.AdjList = [Node('')]
        self.add_keywords(keywords)
        self.set_fail_transitions()

    def find_next_state(self, current_state, value):
        for node in self.AdjList[current_state].next_states:
            if self.AdjList[node].value == value:
                return node
        return None

    def add_keywords(self, keywords):
        """ add all keywords in list of keywords """
        for keyword in keywords:
            current_state = 0
            newNode = False
            for car in keyword.lower():
                if not newNode:
                    child = self.find_next_state(current_state, car)
                if newNode or not child:
                    newNode = True
                    self.AdjList.append(Node(car))
                    child = len(self.AdjList) - 1
                    self.AdjList[current_state].next_states.append(child)
                current_state = child
            self.AdjList[current_state].output.append(keyword)

    def set_fail_transitions(self):
        q = deque()
        for node in self.AdjList[0].next_states:
            q.append(node)
        while q:
            r = q.popleft()
            for child in self.AdjList[r].next_states:
                q.append(child)
                state = self.AdjList[r].fail_state
                while self.find_next_state(state, self.AdjList[child].value) == None and state != 0:
                    state = self.AdjList[state].fail_state
                self.AdjList[child].fail_state = self.find_next_state(state, self.AdjList[child].value)
                if self.AdjList[child].fail_state is None:
                    self.AdjList[child].fail_state = 0
                self.AdjList[child].output = self.AdjList[child].output + self.AdjList[self.AdjList[child].fail_state].output

    def search(self, line):
        """ returns true if line contains any keywords in trie """
        line = line.lower()
        current_state = 0
        keywords_found = []

        for i in range(len(line)):
            while self.find_next_state(current_state, line[i]) is None and current_state != 0:
                current_state = self.AdjList[current_state].fail_state
            current_state = self.find_next_state(current_state, line[i])
            if current_state is None:
                current_state = 0
            else:
                for j in self.AdjList[current_state].output:
                    keywords_found.append((i - len(j) + 1, j))
        return keywords_found

def read_input(input: str) -> tuple:
    patterns, designs = input.split('\n\n')
    patterns = patterns.split(', ')
    designs = [design for design in designs.split('\n') if design]
    return patterns, designs

def puzzle1(input: str) -> int:
    patterns, designs = read_input(input)
    print(f'patterns: {patterns}')
    print(f'designs: {designs}')
    ac = AhoCorasick(patterns)
    matching = 0
    for design in designs:
        result = ac.search(design)
        found = [0] * len(design)
        for start, word in result:
            for i in range(start, start + len(word)):
                found[i] = 1
        if 0 not in found:
            matching += 1

    print(f'matching: {matching}')
    return matching

def find_combinations(length: int, string_list: dict) -> int:

    def backtrack(index):
        result = 0
        for string_length in string_list[index]:
            if index + string_length == length:
                result += 1
            else:
                result += backtrack(index + string_length)
        return result

    return backtrack(0)

def puzzle2(input: str) -> int:
    patterns, designs = read_input(input)
    print(f'patterns: {patterns}')
    print(f'designs: {designs}')
    ac = AhoCorasick(patterns)
    combinations = 0
    for design in designs:
        result = ac.search(design)
        found = [0] * len(design)
        for start, word in result:
            for i in range(start, start + len(word)):
                found[i] = 1
        if 0 not in found:
            words = defaultdict(list)
            for start, word in result:
                words[start].append(len(word))
            combinations += find_combinations(len(design), words)


    print(f'combinations: {combinations}')
    return combinations

input_demo_1 = '''\
r, wr, b, g, bwu, rb, gb, br

brwrr
bggr
gbbr
rrbgbr
ubwu
bwurrg
brgr
bbrgwb
'''

assert puzzle1(input_demo_1) == 6
assert puzzle2(input_demo_1) == 16

input = readInput(2024, 19)
result = puzzle1(input)
# Day 19, puzzle 1: 317
print(f"Puzzle 1: {result}")
result = puzzle2(input)
# Day 19, puzzle 2: ???
print(f"Puzzle 2: {result}")
````
