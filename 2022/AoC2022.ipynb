Rock Paper Scissors is a game between two players. Each game contains many rounds; in each round, the players each simultaneously choose one of Rock, Paper, or Scissors using a hand shape. Then, a winner for that round is selected: Rock defeats Scissors, Scissors defeats Paper, and Paper defeats Rock. If both players choose the same shape, the round instead ends in a draw.

Appreciative of your help yesterday, one Elf gives you an encrypted strategy guide (your puzzle input) that they say will be sure to help you win. The winner of the whole tournament is the player with the highest score. Your total score is the sum of your scores for each round. The score for a single round is the score for the shape you selected (1 for Rock, 2 for Paper, and 3 for Scissors) plus the score for the outcome of the round (0 if you lost, 3 if the round was a draw, and 6 if you won).

Since you can't be sure if the Elf is trying to help you or trick you, you should calculate the score you would get if you were to follow the strategy guide.

The total score is still calculated in the same way, but now you need to figure out what shape to choose so the round ends as indicated. The example above now goes like this:

* In the first round, your opponent will choose Rock (A), and you need the round to end in a draw (Y), so you also choose Rock. This gives you a score of 1 + 3 = 4.
* In the second round, your opponent will choose Paper (B), and you choose Rock so you lose (X) with a score of 1 + 0 = 1.

````python
#day2 => initializer-cell
import re

ROCK = 1
PAPER = 2
SCISSORS = 3

WIN = 6
DRAW = 3
LOSS = 0

MAPPINGS = {
    'A': ROCK,
    'B': PAPER,
    'C': SCISSORS,
    'X': ROCK,
    'Y': PAPER,
    'Z': SCISSORS
}

BATTLE_MATRIX = {
    ROCK: {
        ROCK: DRAW,
        PAPER: LOSS,
        SCISSORS: WIN},
    PAPER: {
        ROCK: WIN,
        PAPER: DRAW,
        SCISSORS: LOSS},
    SCISSORS: {
        ROCK: LOSS,
        PAPER: WIN,
        SCISSORS: DRAW}
}

MAPPINGS_PART2 = {
    'A': ROCK,
    'B': PAPER,
    'C': SCISSORS,
    'X': LOSS,
    'Y': DRAW,
    'Z': WIN
}

BATTLE_MATRIX_PART2 = {
    ROCK: {
        LOSS: SCISSORS,
        DRAW: ROCK,
        WIN: PAPER},
    PAPER: {
        LOSS: ROCK,
        DRAW: PAPER,
        WIN: SCISSORS},
    SCISSORS: {
        LOSS: PAPER,
        DRAW: SCISSORS,
        WIN: ROCK}
}

def read_input(input: str) -> list:
    instrs = [instr.split() for instr in input.split('\n') if instr != '']
    return instrs

def puzzle1(input: str) -> int:
    instrs = read_input(input)
    score = 0
    for instr in instrs:
        opponent = MAPPINGS[instr[0]]
        me = MAPPINGS[instr[1]]
        score += me + BATTLE_MATRIX[opponent][me]
    return score

def puzzle2(input: str) -> int:
    instrs = read_input(input)
    score = 0
    for instr in instrs:
        opponent = MAPPINGS_PART2[instr[0]]
        me = MAPPINGS_PART2[instr[1]]
        score += me + BATTLE_MATRIX_PART2[opponent][me]
    return score

input_demo_1 = '''\
A Y
B X
C Z
'''
assert puzzle1(input_demo_1) == 15
assert puzzle2(input_demo_1) == 12

input = readInput(2022, 2)
score = puzzle1(input)
# Day 2, puzzle 1: 15691
print(f"Puzzle 1: Your score are rock, papper, scissor: {score}")
score = puzzle2(input)
# Day 2, puzzle 1: 12989
print(f"Puzzle 2: Your score are rock, papper, scissor: {score}")
````

One Elf has the important job of loading all of the rucksacks with supplies for the jungle journey. Unfortunately, that Elf didn't quite follow the packing instructions, and so a few items now need to be rearranged.

Each rucksack has two large compartments. All items of a given type are meant to go into exactly one of the two compartments. The Elf that did the packing failed to follow this rule for exactly one item type per rucksack.

The Elves have made a list of all of the items currently in each rucksack (your puzzle input), but they need your help finding the errors. Every item type is identified by a single lowercase or uppercase letter (that is, a and A refer to different types of items).

The list of items for each rucksack is given as characters all on a single line. # Part Two

As you finish identifying the misplaced items, the Elves come to you with another issue.

For safety, the Elves are divided into groups of three. Every Elf carries a badge that identifies their group. For efficiency, within each group of three Elves, the badge is the only item type carried by all three Elves. That is, if a group's badge is item type B, then all three Elves will have item type B somewhere in their rucksack, and at most two of the Elves will be carrying any other item type.

The problem is that someone forgot to put this year's updated authenticity sticker on the badges. All of the badges need to be pulled out of the rucksacks so the new authenticity stickers can be attached.

Additionally, nobody wrote down which item type corresponds to each group's badges. The only way to tell which item type is the right one is by finding the one item type that is common between all three Elves in each group.

Every set of three lines in your list corresponds to a single group, but each group can have a different badge item type. So, in the above example, the first group's rucksacks are the first three lines:
```
vJrwpWtwJgWrhcsFMMfFFhFp
jqHRNqRjqzjGDLGLrsFMfFZSrLrFZsSL
PmmdzqPrVvPwwTWBwg
```
And the second group's rucksacks are the next three lines:
```
wMqvLMZHhHMvwLHjbvcjnnSBnvTQFn
ttgJtRGJQctTZtZT
CrZsJsPPZsGzwwsLwLmpwMDw
```
In the first group, the only item type that appears in all three rucksacks is lowercase r; this must be their badges. In the second group, their badge item type must be Z.

Priorities for these items must still be found to organize the sticker attachment efforts: here, they are 18 (r) for the first group and 52 (Z) for the second group. The sum of these is 70.

Find the item type that corresponds to the badges of each three-Elf group. What is the sum of the priorities of those item types?

Your puzzle answer was 2607.

Both parts of this puzzle are complete! They provide two gold stars: **

````python
#day3 => initializer-cell
import re

def read_input(input: str) -> list:
    rucksacks = [rucksack for rucksack in input.split('\n') if rucksack != '']
    return rucksacks

def get_common_item(rucksack: str) -> str:
    compartment1 = set(rucksack[:len(rucksack)//2])
    compartment2 = set(rucksack[len(rucksack)//2:])
    return list(compartment1.intersection(compartment2))[0]

def get_priority(item: str) -> int:
    if item.islower():
        return ord(item) - ord('a') + 1
    else:
        return ord(item) - ord('A') + 27

def puzzle1(input: str) -> int:
    rucksacks = read_input(input)
    return sum([get_priority(get_common_item(rucksack)) for rucksack in rucksacks])

def puzzle2(input: str) -> int:
    rucksacks = read_input(input)
    # loop through rucksacks by block of 3 lines
    total_priority = 0
    for i in range(0, len(rucksacks), 3):
        group = rucksacks[i:i+3]
        common_item = list(set(group[0]).intersection(group[1]).intersection(group[2]))[0]
        total_priority += get_priority(common_item)
    return total_priority

input_demo_1 = '''\
vJrwpWtwJgWrhcsFMMfFFhFp
jqHRNqRjqzjGDLGLrsFMfFZSrLrFZsSL
PmmdzqPrVvPwwTWBwg
wMqvLMZHhHMvwLHjbvcjnnSBnvTQFn
ttgJtRGJQctTZtZT
CrZsJsPPZsGzwwsLwLmpwMDw
'''
assert puzzle1(input_demo_1) == 157
assert puzzle2(input_demo_1) == 70

input = readInput(2022, 3)
priority_sum = puzzle1(input)
# Day 2, puzzle 1: 7597
print(f"Puzzle 1: Your priority sum: {priority_sum}")
priority_sum = puzzle2(input)
# Day 2, puzzle 2: 2607
print(f"Puzzle 2: Your priority sum: {priority_sum}")
````

````python
#day4 => initializer-cell
import re

RE_PAIRS = re.compile(r'(\d+)-(\d+),(\d+)-(\d+)')

def read_input(input: str) -> list:
    pairs = [list(map(int, RE_PAIRS.match(line).groups())) for line in input.split('\n') if line != '']
    return pairs

def puzzle1(input: str) -> int:
    pairs = read_input(input)
    count = 0
    for pair in pairs:
        if (pair[0] <= pair[2] and pair[1] >= pair[3]) or (pair[0] >= pair[2] and pair[1] <= pair[3]):
            count += 1
    print('count:', count)
    return count

def puzzle2(input: str) -> int:
    pairs = read_input(input)
    count = 0
    for pair in pairs:
        if (pair[0] >= pair[2] and pair[0] <= pair[3]) or (
            (pair[1] >= pair[2] and pair[1] <= pair[3]) or (
                (pair[0] <= pair[2] and pair[1] >= pair[3]) or (
                    (pair[0] >= pair[2] and pair[1] <= pair[3])))):
            count += 1
    print('count:', count)
    return count

input_demo_1 = '''\
2-4,6-8
2-3,4-5
5-7,7-9
2-8,3-7
6-6,4-6
2-6,4-8
'''
assert puzzle1(input_demo_1) == 2
assert puzzle2(input_demo_1) == 4

input = readInput(2022, 4)
assignments = puzzle1(input)
# Day 2, puzzle 1: 7597
print(f"Puzzle 1: Your priority sum: {assignments}")
assignments = puzzle2(input)
# Day 2, puzzle 2: 852
print(f"Puzzle 2: Your priority sum: {assignments}")
````

````python
#day5 => initializer-cell
import re

RE_STACK = re.compile(r'\[([A-Z])\]')
RE_MOVE = re.compile(r'move (\d+) from (\d+) to (\d+)')

def read_input(input: str) -> list:
    stacks_part, moves_part = input.split('\n\n')

    stacks_tab = stacks_part.split('\n')[::-1][1:]
    stack_count = (len(stacks_tab[-1]) + 1) // 4
    stacks = [[] for _ in range(stack_count)]
    for stack in stacks_tab:
        for id, i in enumerate(range(0, len(stack) - 1, 4)):
            if stack[i:i+3] == '   ':
                continue
            stacks[id].append(stack[i+1])
    moves = [list(map(int, RE_MOVE.match(move).groups())) for move in moves_part.split('\n') if move != '']

    return stacks, moves


def puzzle1(input: str) -> str:
    stacks, moves = read_input(input)
    for num, src, dst in moves:
        for _ in range(num):
            stacks[dst-1].append(stacks[src-1].pop())
    return ''.join([stack[-1] for stack in stacks])

def puzzle2(input: str) -> str:
    stacks, moves = read_input(input)
    for num, src, dst in moves:
        stacks[dst-1].extend(stacks[src-1][-num:])
        stacks[src-1] = stacks[src-1][:-num]
    return ''.join([stack[-1] for stack in stacks])

input_demo_1 = '''\
    [D]
[N] [C]
[Z] [M] [P]
 1   2   3

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2
'''
assert puzzle1(input_demo_1) == 'CMZ'
assert puzzle2(input_demo_1) == 'MCD'

input = readInput(2022, 5)
result = puzzle1(input)
# Day 2, puzzle 1: VGBBJCRMN
print(f"Puzzle 1: Your code is : {result}")
result = puzzle2(input)
# Day 2, puzzle 2: LBBVJBRMH
print(f"Puzzle 2: Your code is : {result}")
````

````python
#day7 => initializer-cell

def get_file_system_from_commands(commands: list) -> dict:
    current_path = ['root']
    mode = 'wait'
    file_system = {
        'size': 0,
        'children': {
            'root': {
                'size': 0,
                'children': {}
            }
        }
    }
    for command in commands:
        if command.startswith('$ cd'):
            mode = 'wait'
            if command == '$ cd /':
                current_path = ['root']
            elif command == '$ cd ..':
                current_path.pop()
            else:
                current_path.append(command.split(' ')[2])
        elif command.startswith('$ ls'):
            mode = 'ls'
            continue
        elif mode == 'ls':
            if command.startswith('dir'):
                dir_name = command.split(' ')[1]
                active_folder = file_system
                for path in current_path:
                    active_folder = active_folder['children'][path]
                active_folder['children'][dir_name] = {
                    'size': 0,
                    'children': {}
                }
                continue
            else:
                size, file_name = command.split(' ')
                active_folder = file_system
                for path in current_path:
                    active_folder = active_folder['children'][path]
                active_folder['children'][file_name] = {
                    'size': int(size)
                }
            continue
    return file_system

def get_folder_size(folder):
    size = 0
    for child in folder['children'].values():
        if not 'children' in child:
            size += child['size']
        else:
            size += get_folder_size(child)
    folder['size'] = size
    return size

def flatten_file_system(file_system: dict, folder_sizes: dict, current_path: str="") -> None:
    """
    Flattens the file_system data structure and stores folder sizes in folder_sizes.

    Args:
        file_system (dict): The file system data structure.
        folder_sizes (dict): A dictionary to store folder sizes.
        current_path (str): The current path during traversal.
    """

    for name, child in file_system['children'].items():
        path = current_path + "/" + name if current_path else name
        if 'children' in child:  # It's a directory
            folder_sizes[path] = child['size']
            flatten_file_system(child, folder_sizes, path)

def puzzle1(input: str) -> int:
    commands = [command for command in input.split('\n') if command]
    file_system = get_file_system_from_commands(commands)
    active_folder = file_system
    get_folder_size(active_folder)

    folder_sizes = {}
    flatten_file_system(file_system, folder_sizes)
    less_than_100k = [size for size in folder_sizes.values() if size <= 100000]

    return sum(less_than_100k)

def puzzle2(input: str) -> int:
    commands = [command for command in input.split('\n') if command]
    file_system = get_file_system_from_commands(commands)
    active_folder = file_system
    get_folder_size(active_folder)

    total_space_disk = 70000000
    needed_space = 30000000
    folder_sizes = {}
    flatten_file_system(file_system, folder_sizes)
    current_free_disk_space = total_space_disk - folder_sizes['root']
    space_to_free = needed_space - current_free_disk_space
    folder_sizes = {name: size for name, size in folder_sizes.items() if size >= space_to_free}
    return min(folder_sizes.values())


input_demo_1 = '''\
$ cd /
$ ls
dir a
14848514 b.txt
8504156 c.dat
dir d
$ cd a
$ ls
dir e
29116 f
2557 g
62596 h.lst
$ cd e
$ ls
584 i
$ cd ..
$ cd ..
$ cd d
$ ls
4060174 j
8033020 d.log
5626152 d.ext
7214296 k
'''
assert puzzle1(input_demo_1) == 95437
assert puzzle2(input_demo_1) == 24933642

input = readInput(2022, 7)
result = puzzle1(input)
# Day 2, puzzle 1: 1086293
print(f"Puzzle 1: sum of folders under 100K : {result}")
result = puzzle2(input)
# Day 2, puzzle 2: 366028
print(f"Puzzle 2: size of folder to delete : {result}")
````
